```python
seg_type = pd.get_dummies(file['Segment type'])
seg_type = seg_type.drop('gender',axis=1)
seg_type
```

```python
seg_dcpt = pd.get_dummies(file['Segment Description'])
seg_dcpt = seg_dcpt.drop('Appalachian State University',axis=1)
seg_dcpt
```

```python
ans = pd.get_dummies(file['Answer'])
ans = ans.drop("I don't use Tinder",axis=1)
ans
```

```python
file = file.drop(['Segment Description', 'Answer', 'Segment type'], axis=1)
file = pd.concat([file,seg_type,seg_dcpt,ans], axis=1)
```

```python
file #Converted every string into numerical form by using dummy variables
```

```python
#3. Spilt Data in 2 parts
X_final=file.drop('It became a relationship',axis =1)
y_final=file['It became a relationship']
```

```python
X_final.shape
```

```python
y_final.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_final, y_final, test_size=0.2)
```

## Model_1

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```

### Model_1_a

```python
model_1_a = DecisionTreeClassifier(random_state=0,max_depth=30)

model_1_a = model_1_a.fit(X_train,y_train)

y_pred_1_a = model_1_a.predict(X_test)

confusion_matrix(y_test,y_pred_1_a)
```

```python

accuracy_score(y_test,y_pred_1_a)
```

### Model_1_b

```python
model_1_b = DecisionTreeClassifier(random_state=1,max_depth=60)

model_1_b = model_1_b.fit(X_train,y_train)

y_pred_1_b = model_1_b.predict(X_test)

confusion_matrix(y_test,y_pred_1_b)
```

```python
accuracy_score(y_test,y_pred_1_b)
```

### Model_1_c

```python
model_1_c = DecisionTreeClassifier(random_state=42,max_depth=90)

model_1_c = model_1_c.fit(X_train,y_train)

y_pred_1_c = model_1_c.predict(X_test)

confusion_matrix(y_test,y_pred_1_c)
```

```python
accuracy_score(y_test,y_pred_1_c)
```

## Model_2

```python
from sklearn.ensemble import RandomForestClassifier
```

### Model_2_a

```python
model_2_a = RandomForestClassifier(n_estimators=20, criterion="gini",  n_jobs=-1, random_state=101, 
                                   max_features = 154, min_samples_leaf = 10)#n_jobs=-1 means all processor and max_feature = 154 (total features in new data set)
model_2_a = model_2_a.fit(X_train,y_train)

y_pred_2_a = model_2_a.predict(X_test)

confusion_matrix(y_test,y_pred_2_a)
```

```python
accuracy_score(y_test,y_pred_2_a)
```

### Model_2_b

```python
model_2_b = RandomForestClassifier(n_estimators=50, criterion="entropy", n_jobs=-1, random_state=1, 
                                   max_features = 154, min_samples_leaf = 20) #n_jobs=-1 means all processor and max_feature = 154 (total features in new data set)
model_2_b = model_2_b.fit(X_train,y_train)

y_pred_2_b = model_2_b.predict(X_test)

confusion_matrix(y_test,y_pred_2_b)
```

```python
accuracy_score(y_test,y_pred_2_b)
```
